```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Dense, Embedding, Add, Dot, Softmax, Activation, Flatten, Concatenate
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import plot_model

# Directories
base_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence'
data_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence\data'
models_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence\models'
pre_processed_data_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence\pre_processed_data'

def scaled_dot_product_attention(query, key, value):
    """Calculate scaled dot-product attention."""
    attention_scores = Dot(axes=[2, 2])([query, key])
    attention_scores_scaled = Activation('softmax')(attention_scores)
    context_vector = Dot(axes=[2, 1])([attention_scores_scaled, value])
    return context_vector

# Custom Model with Attention
def build_attention_model(vocab_size, max_len):
    # Image input
    image_input = Input(shape=(2048,))  # 2048 features from InceptionV3
    image_dense = Dense(256, activation='relu')(image_input)

    # Sequence input
    seq_input = Input(shape=(max_len,))
    seq_embedding = Embedding(vocab_size, 256, mask_zero=True)(seq_input)

    # Attention mechanism
    query = Dense(256)(seq_embedding)
    key = Dense(256)(image_dense)
    value = Dense(256)(image_dense)

    # Expand image features to match sequence dimensions
    image_features_expanded = Flatten()(Dense(max_len, activation='relu')(image_dense))
    image_features_expanded = Dense(256)(image_features_expanded)

    context_vector = scaled_dot_product_attention(query, key, value)
    combined = Concatenate()([context_vector, seq_embedding])

    # Fully connected output layer
    dense_1 = Dense(256, activation='relu')(combined)
    output = Dense(vocab_size, activation='softmax')(dense_1)

    model = Model(inputs=[image_input, seq_input], outputs=output)
    model.compile(optimizer='adam', loss='categorical_crossentropy')
    return model

# Load data
train_captions = os.path.join(pre_processed_data_dir, 'train_captions.txt')
test_captions = os.path.join(pre_processed_data_dir, 'test_captions.txt')
train_img_dir = os.path.join(data_dir, 'train')
test_img_dir = os.path.join(data_dir, 'test')

train_image_features = pickle.load(open(os.path.join(pre_processed_data_dir, 'train_image_features.pkl'), 'rb'))
test_image_features = pickle.load(open(os.path.join(pre_processed_data_dir, 'test_image_features.pkl'), 'rb'))

import image_captioning_utils as icu

train_image_ids = list(train_image_features.keys())
train_processor = icu.CaptionProcessor(train_captions, train_image_ids)

VOCAB_SIZE = len(train_processor.tokenizer.word_index) + 1
MAX_LEN = train_processor.max_len

# Create sequences for training
X1, X2, y = train_processor.create_sequences(train_image_features, VOCAB_SIZE)

# Build and train the model
attention_model = build_attention_model(VOCAB_SIZE, MAX_LEN)
attention_model.summary()

history = attention_model.fit([X1, X2], y, epochs=10, batch_size=64, validation_split=0.2)

def plot_history(history):
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(len(loss))

    plt.figure(figsize=(20, 5))
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')

    plt.show()

plot_history(history)

# Save the model
attention_model.save(os.path.join(models_dir, 'attention_model.h5'))
loaded_model = load_model(os.path.join(models_dir, 'attention_model.h5'))

# Generate caption for a new image
random_id = np.random.choice(list(train_processor.captions_dict.keys()))
cap_id = random_id
train_image_path = os.path.join(train_img_dir, cap_id + '.jpg')

extractor = icu.ImageFeatureExtractor()
train_image_feature = extractor.extract_features(train_image_path)
train_image_feature = np.squeeze(train_image_feature)

generated_caption = icu.generate_caption(
    loaded_model, np.array([train_image_feature]), train_processor.tokenizer, MAX_LEN
)

print("Actual Caption: ----------------------------")
for cap in train_processor.captions_dict[cap_id]:
    print(cap)
print("--------------------------------------------")

print("Generated Caption: ", generated_caption)

# Display the image
img = Image.open(train_image_path)
plt.imshow(img)
plt.show()
```

```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Dense, Embedding, Add, MultiHeadAttention, LayerNormalization, Dropout, Flatten, Reshape
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import plot_model
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# Directories
base_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence'
data_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence\data'
models_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence\models'
pre_processed_data_dir = r'C:\Users\UTKARSH\Desktop\visual intelligence\pre_processed_data'

# Transformer Model
class TransformerModel:
    def __init__(self, vocab_size, max_len, embedding_dim=256, num_heads=8, ff_dim=512, dropout_rate=0.1):
        self.vocab_size = vocab_size
        self.max_len = max_len
        self.embedding_dim = embedding_dim
        self.num_heads = num_heads
        self.ff_dim = ff_dim
        self.dropout_rate = dropout_rate

    def transformer_encoder(self, inputs):
        attention_output = MultiHeadAttention(
            num_heads=self.num_heads, key_dim=self.embedding_dim
        )(inputs, inputs)
        attention_output = Dropout(self.dropout_rate)(attention_output)
        attention_output = Add()([inputs, attention_output])
        attention_output = LayerNormalization(epsilon=1e-6)(attention_output)

        ff_output = Dense(self.ff_dim, activation="relu")(attention_output)
        ff_output = Dense(self.embedding_dim)(ff_output)
        ff_output = Dropout(self.dropout_rate)(ff_output)
        sequence_output = Add()([attention_output, ff_output])
        sequence_output = LayerNormalization(epsilon=1e-6)(sequence_output)
        return sequence_output

    def build_model(self):
        # Image input
        image_input = Input(shape=(2048,))
        image_dense = Dense(self.embedding_dim, activation="relu")(image_input)
        image_reshaped = Reshape((1, self.embedding_dim))(image_dense)

        # Sequence input
        seq_input = Input(shape=(self.max_len,))
        seq_embedding = Embedding(self.vocab_size, self.embedding_dim, mask_zero=True)(seq_input)

        # Transformer Encoder
        transformer_output = self.transformer_encoder(seq_embedding)

        # Combine image features with sequence features
        combined = Add()([image_reshaped, transformer_output])

        # Flatten and fully connected output layer
        flattened = Flatten()(combined)
        dense_1 = Dense(self.embedding_dim, activation="relu")(flattened)
        output = Dense(self.vocab_size, activation="softmax")(dense_1)

        model = Model(inputs=[image_input, seq_input], outputs=output)
        model.compile(optimizer="adam", loss="categorical_crossentropy")
        return model

# Load data
train_captions = os.path.join(pre_processed_data_dir, 'train_captions.txt')
test_captions = os.path.join(pre_processed_data_dir, 'test_captions.txt')
train_img_dir = os.path.join(data_dir, 'train')
test_img_dir = os.path.join(data_dir, 'test')

train_image_features = pickle.load(open(os.path.join(pre_processed_data_dir, 'train_image_features.pkl'), 'rb'))
test_image_features = pickle.load(open(os.path.join(pre_processed_data_dir, 'test_image_features.pkl'), 'rb'))

import image_captioning_utils as icu

train_image_ids = list(train_image_features.keys())
train_processor = icu.CaptionProcessor(train_captions, train_image_ids)

test_image_ids = list(test_image_features.keys())
test_processor = icu.CaptionProcessor(test_captions, test_image_ids)

VOCAB_SIZE = len(train_processor.tokenizer.word_index) + 1
MAX_LEN = train_processor.max_len

# Create sequences for training
X1, X2, y = train_processor.create_sequences(train_image_features, VOCAB_SIZE)

# Build and train the model
transformer = TransformerModel(VOCAB_SIZE, MAX_LEN)
transformer_model = transformer.build_model()
transformer_model.summary()

history = transformer_model.fit([X1, X2], y, epochs=10, batch_size=64, validation_split=0.2)

def plot_history(history):
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(len(loss))

    plt.figure(figsize=(20, 5))
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')

    plt.show()

plot_history(history)

# Save the model
transformer_model.save(os.path.join(models_dir, 'transformer_model.h5'))
loaded_model = load_model(os.path.join(models_dir, 'transformer_model.h5'))

# Generate caption for a new image
random_id = np.random.choice(list(test_processor.captions_dict.keys()))
cap_id = random_id
test_image_path = os.path.join(test_img_dir, cap_id + '.jpg')

extractor = icu.ImageFeatureExtractor()
test_image_feature = extractor.extract_features(test_image_path)
test_image_feature = np.squeeze(test_image_feature)

generated_caption = icu.generate_caption(
    loaded_model, np.array([test_image_feature]), test_processor.tokenizer, MAX_LEN
)

print("Actual Caption: ----------------------------")
actual_captions = test_processor.captions_dict[cap_id]
for cap in actual_captions:
    print(cap)
print("--------------------------------------------")

print("Generated Caption: ", generated_caption)
# Display the image
img = Image.open(test_image_path)
plt.imshow(img)
plt.show()
```
